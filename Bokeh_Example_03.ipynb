```python
from scipy.misc import lena
import bokeh.plotting as bp
from bokeh.resources import CDN
from bokeh.embed import file_html

import numpy

from IPython.display import display, HTML
```

```python
lena_img = lena()/256.0

plot = bp.figure(plot_width=lena_img.shape[0], plot_height=lena_img.shape[1], 
                 x_range=[0, lena_img.shape[0]], y_range=[0, lena_img.shape[1]])
plot.image(image=[numpy.flipud(lena_img)], x=[0], y=[0],
           dw=[lena_img.shape[0]], dh=[lena_img.shape[1]])

html = file_html(plot, CDN, "my plot")

display(HTML(html))
```
